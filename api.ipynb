An API stands for Application Processing interface which is mainly used to process the input/output across different platforms which can be made from same or different technologies,it acts as translator. For eg while logging in to our pw website for accessing the course we enter our credentials on website which is further verfified from servers which consists of this data.

Q2:
Advantages:
Transfer of data can be done across different applications/web pages which further eases the task of developer to don't start any work from scratch.
Disadvantages:
As a single point of entry, an API is a gateway and can become a hacker's primary target. Once the API is compromised, all other applications and systems become vulnerable.
